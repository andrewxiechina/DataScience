### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo["price"] = chipo["item_price"].apply(lambda x: float(x[1:-1]))
chipo[chipo["price"] > 10.00].item_name.value_counts().count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.drop_duplicates(["item_name"]).sort_values("price", ascending=False)[["item_name", "item_price"]]
```

### Step 6. Sort by the name of the item

```python
chipo.drop_duplicates(["item_name"]).sort_values("item_name").item_name
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
item = chipo.drop_duplicates(["item_name"]).sort_values("price", ascending=False)["item_name"].values[0]
chipo[chipo["item_name"] == item].quantity.sum()
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
item = "Veggie Salad Bowl"
chipo[chipo["item_name"] == item].quantity.sum()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
item = "Canned Soda"
temp = chipo[chipo["item_name"] == item]
temp[temp["quantity"] > 1].order_id.count()
```
